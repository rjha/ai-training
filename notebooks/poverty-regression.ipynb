```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

"""
Show linear regression with scatter plot using the seaborne library 
The point of this exercise is to show how easy it is to do linear 
regression with a modern stack. 

The dataset is a relationship between poverty and teenage births 
The states with higher poverty generally have higher teenage pregnancy. 

STEP 1: show original data points and fit using seaborne
Load the dataset from penn state stat 462 
data is taken from https://online.stat.psu.edu/stat462/node/101/
"""

poverty_df = pd.read_csv('/Users/rjha/code/github/ai-training/data/psu/poverty.dat', sep=r'\s+')
print(poverty_df.head())

```

```python

# 2. Create the scatter plot with the regression line
sns.regplot(
    data=poverty_df,
    x="PovPct",
    y="Brth15to17",
    scatter_kws={"color": "royalblue", "alpha": 0.7, "s": 50},
    line_kws={"color": "crimson", "linewidth": 2},
)

# 4. Add clear labels and title
plt.title(
    "Teen Birth Rate vs. Poverty Level by US State", fontsize=14, pad=15
)
plt.xlabel("Poverty Rate (%)", fontsize=12)
plt.ylabel("Birth Rate (per 1000 females aged 15-17)", fontsize=12)

# 5. Display the visualization
plt.tight_layout()
plt.show()
```
