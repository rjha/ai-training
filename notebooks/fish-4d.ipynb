```python
import pandas as pd
import matplotlib.pyplot as plt
from random import randint

# show 4D scatter plot using pandas and fish data 
# 1. Load the dataset (replace with your local file path if needed)
fish_df = pd.read_csv('/Users/rjha/code/github/ai-training/data/fish-data.csv')
print(fish_df.head())


# colors = ['red' if val == 'Bream' else 'blue' for val in fish_df['Species']]
# random_color="#" + f"{randint(0, 0xFFFFFF):06x}"
# Do not use both c and cmap options for STRING categories like species
# just pass c='Species' and no cmap.
# You can use both c and cmap options for numeric categories like CAT 
# 

fish_df.plot.scatter(
    x='Length1',
    y='Height',
    s='Weight',                # Map 3rd dimension to size
    c='CAT',               # Map 4th dimension to color
    cmap='viridis',           # Choose color palette
    alpha=0.6,                 # Set transparency for overlapping bubbles
    figsize=(10, 6),
    title="Fish 4D Scatter Plot in Pandas"
)

plt.show()

```
