```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

"""
Demo to show plots of multivariate data using pandas parallel coordinates 
STEP 1. Load the data into a pandas data frame 

"""
fish_df = pd.read_csv('/Users/rjha/code/github/ai-training/data/fish-data.csv')
print(fish_df.head())


```

```python
# STEP 2: Create the parallel coordinates plot without filtering data 
# This will show all features with their numeric value across species 

plt.figure(figsize=(12, 7))
parallel_coordinates(fish_df, class_column='Species', colormap='inferno')

# 4. Customize the plot
plt.title('Parallel Coordinates of Fish Market Data', fontsize=16)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(loc='upper right')
plt.show()
```

```python
# STEP3: plot a filtered data frame 
# Filter the columns first, keeping the class column
weight_length1_df = fish_df[["Weight", "Length1", "Species"]]
parallel_coordinates(weight_length1_df, class_column="Species", colormap="inferno")
```

```python
fish_df.plot.scatter(
    x='Length1',
    y='Height',
    s='Weight',                # Map 3rd dimension to size
    c='CAT',               # Map 4th dimension to color
    cmap='viridis',           # Choose color palette
    alpha=0.6,                 # Set transparency for overlapping bubbles
    figsize=(10, 6),
    title="Fish 4D Scatter Plot in Pandas"
)

plt.show()
```
