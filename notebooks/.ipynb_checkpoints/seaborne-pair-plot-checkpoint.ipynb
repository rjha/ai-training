```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# show seaborn pair plot on fish data 
# 1. Load the dataset (replace with your local file path if needed)
fish_df = pd.read_csv('./Fish-data.csv')
print(fish_df.head())
print(fish_df.dtypes)



```

```python
sns.displot(
  data=fish_df,
  x="Weight",
  hue="Species",
  kind="hist",
  height=6,
  aspect=1.4,
  bins=15
)
plt.show()
```

```python
sns.pairplot(fish_df, kind='scatter', hue='Species');
```
