```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import parallel_coordinates

"""

The Iris dataset is a classic benchmarking dataset by British statistician 
Ronald Fisher in 1936. It records the physical dimensions of 150 total flower 
samples across three different species of the Iris flower. 

Iris setosa is completely isolated from the other two species based on its 
small size, making it a perfect example of an easy classification task.

"""

# 1. Load a multivariate dataset (e.g., Iris dataset)
df = sns.load_dataset("iris")

# 2. Generate the parallel coordinates plot
# 'species' is the categorical column used for color coding
parallel_coordinates(df, "species", colormap="Set2")

# 3. Display the visualization
plt.title("Parallel Coordinates Plot - Iris Dataset")
plt.show()
```

```python
"""
We can see that petal measurements are much better at classification 
than sepal measurements. 

"""
# Filter the columns first, keeping the class column
petal_df = df[["petal_length", "petal_width", "species"]]

# Plot without needing the 'cols' argument
parallel_coordinates(petal_df, class_column="species", colormap="Set2")
```
