```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix


# 1. Load the dataset (replace with your local file path if needed)
fish_df = pd.read_csv('./Fish-data.csv')
print(fish_df.head())


# Plot a matrix of all dimensional pairs
scatter_matrix(
    fish_df, 
    alpha=0.5, 
    figsize=(10, 10), 
    diagonal='kde' # Displays Kernel Density Estimation instead of histograms
)

plt.show()

```
